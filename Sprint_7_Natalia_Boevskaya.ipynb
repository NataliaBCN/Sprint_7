**Ejercicio 1**  Calculadora del índice de masa corporal (IMC) 

```python
def bmi(weight, height_cm):
    """This function takes weight and height as input and calculate Body Mass Index (BMI)"""
    height_m = height_cm / 100   # Convert cm to meters
    return weight / (height_m ** 2)

def bmi_type(bmi):
    """This function takes BMI and classify it """
    if bmi < 5:
        return "It seems that the input was incorrect."
    elif bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 25:
        return "Normal weight"
    elif 25 <= bmi < 30:
        return "Overweight"
    elif 30 <= bmi <= 95:
        return "Obese"    
    else:
        return "It seems that the input was incorrect."
    
def get_float_input(input_measure):
    """Function used to get a valid numeric input """
    while True:
        try:
            float_input = float(input(f'Enter your {input_measure}: '))
            return float_input  
        except ValueError:
            print("Input Error! That is not a valid number. Try again!")

# Solicitar peso
input_measure = "weight (in kg)"
weight = get_float_input(input_measure)
print(f'Your weight is {weight} kg.')


# Solicitar altura en centímetros
input_measure = "height (in cm)"
height_cm = get_float_input(input_measure)
print(f'Your height is {height_cm} cm.')

# Calculate bmi
bmi_value = bmi(weight, height_cm)
print(f'You Body Mass Index is {bmi_value:.2f}')

# Classify bmi
category = bmi_type(bmi_value)
print(f'Body Mass Index classification: {category}')

```

**Ejercicio 2**  
Convertidor de temperaturas  
Existen varias unidades de temperatura utilizadas en diferentes contextos y regiones. Las más comunes son Celsius (°C), Fahrenheit (°F) y Kelvin (K). También existen otras unidades como Rankine (°Ra) y Réaumur (°Re).
Selecciona al menos dos convertidores, de tal forma que al introducir una temperatura, se devuelvan al menos dos conversiones.

[°F] = 9/5 ( [°C]) + 32  
[K] = [°C] + 273  
[°R] = ([°C] + 273.15) × 9/5  
[°C] = 5/9 ([°F] - 32)  
[K] = 5/9 ([°F] - 32) + 273  
[°R] = [°F] + 459.67  
[°F] = 9/5 ([K] - 273) + 32  
[°C] = [K] - 273  
[°R] = [K] × ​9/5  
[°C] = ([°R] − 491.67) × 5/9  
[°F] = [°R] − 459.67  
[K] = [°R] × ​5/9  


```python
def get_float_input(from_unit):
    """Function used to get a valid numeric input """
    while True:
        try:
            value = float(input(f"Enter temperature in {from_unit}: "))
            return value  
        except ValueError:
            print("Input Error! That is not a valid number. Try again!")
            
            
def temperature_converter():
    """Convert temperature to all other units."""
    print("Welcome to the Temperature Converter!")
    print("Units: C (Celsius), F (Fahrenheit), K (Kelvin), Ra (Rankine)")
    
    conversions = {
        ("C", "F"): lambda t: (t * 9/5) + 32,  # Celsius to Fahrenheit
        ("C", "K"): lambda t: t + 273.15,     # Celsius to Kelvin
        ("C", "Ra"): lambda t: (t + 273.15) * 9/5,  # Celsius to Rankine
        ("F", "C"): lambda t: (t - 32) * 5/9,  # Fahrenheit to Celsius
        ("F", "K"): lambda t: (t - 32) * 5/9 + 273.15,  # Fahrenheit to Kelvin
        ("F", "Ra"): lambda t: t + 459.67,  # Fahrenheit to Rankine    
        ("K", "C"): lambda t: t - 273.15,  # Kelvin to Celsius
        ("K", "F"): lambda t: (t - 273.15) * 9/5 + 32,  # Kelvin to Fahrenheit
        ("K", "Ra"): lambda t: t * 9/5,  # Kelvin to Rankine
        ("Ra", "C"): lambda t: (t - 491.67) * 5/9,  # Rankine to Celsius
        ("Ra", "F"): lambda t: t - 459.67,  # Rankine to Fahrenheit
        ("Ra", "K"): lambda t: t * 5/9  # Rankine to Kelvin
    }
    
    # Input: temperature´s unite
    from_unit = input("Enter the temperature unit (C/F/K/R): ").strip()
    if from_unit.lower() == "ra":
        from_unit = "Ra"
    else:
        from_unit = from_unit.upper()
    
    # Input: temperature
    value=get_float_input(from_unit)
    
    print(f"\nConverting {value}°{from_unit} to all other units:\n")
    
    # Perform conversions to all other units
    for to_unit in ["C", "F", "K", "Ra"]:
        if from_unit == to_unit:
            continue  # Skip conversion to the same unit
        if (from_unit, to_unit) in conversions:
            result = conversions[(from_unit, to_unit)](value)
            print(f"{value}°{from_unit} = {result:.2f}°{to_unit}")
        else:
            print(f"Conversion from {from_unit} to {to_unit} not available.")

# Run the temperature converter
temperature_converter()
```

**Ejercicio 3**  
Contador de palabras en un texto  
Escribe una función que, dado un texto, muestre cuántas veces aparece cada palabra.

```python
import string  # To access string.punctuation for removing punctuation

# Taking user input and printing it
text = str(input("Input your text here"))  # Taking user input and converting it to a string
print(text)  # Printing the inputted text

def words_counter(text):
    """This function takes a text as input and counts how many times each word appears in it"""
    
    # Step 1: Remove punctuation and convert the text to lowercase
    table = str.maketrans("", "", string.punctuation) # `string.punctuation` contains all common punctuation marks
    """
    Create a translation table that maps characters to be replaced.
    str.maketrans(x, y, z) has 3 parts:
    - x: a string of characters you want to replace (optional).
    - y: a string of characters that will replace characters in x (optional).
    - z: a string of characters you want to delete (optional).
    """
    cleaned_text = text.translate(table).lower()  # convert to lowercase
    
    print(cleaned_text)  # Printing cleaned text to verify the transformation
    
    # Step 2: Split the cleaned text into a list of words
    words = cleaned_text.split() 
    
    # Step 3: Create a dictionary to count word occurrences
    counter = {}  # Dictionary to store word counts
    
    for word in words:
        if word in counter:  # If the word already exists in the dictionary
            counter[word] += 1  # Increment the count of the word
        else:
            counter[word] = 1  # If the word doesn't exist, add it with count 1
    
    # Step 4: Display the word counts
    for word, quantity in counter.items():
        print(f"'{word}': {quantity} times")  # Print each word and its count

# Call the function to count words in the user-provided text
words_counter(text)
```

**Ejercicio 4**  
Diccionario inverso  
Resulta que el cliente tiene una encuesta muy antigua almacenada en un diccionario y necesita que los resultados estén al revés, es decir, que se intercambien las claves y los valores. Los valores y las claves en el diccionario original son únicos; si este no es el caso, la función deberá imprimir un mensaje de advertencia.

```python
def invert_dictionary(original_dict):
    """This function takes a dictionary as input and inverts values with keys"""
    # Check if all values are unique
    if len(original_dict) != len(set(original_dict.values())):
        print("Warning: The dictionary contains duplicate values.")
        return None  
    
    # Invert the dictionary
    inverted_dict = {v: k for k, v in original_dict.items()}
    
    return inverted_dict

# Example 1:
original_dict = {'a': 1, 'b': 2, 'c': 3}
print("Original dictionary:", original_dict)
inverted_dict = invert_dictionary(original_dict)
print("Inverted dictionary:", inverted_dict)

# Example 2:
original_dict = {'x': 'apple', 'y': 'banana', 'z': 'banana'}
print("Original dictionary:", original_dict)
inverted_dict = invert_dictionary(original_dict)
print("Inverted dictionary:", inverted_dict)
```

Diccionario inverso con duplicados  
Continuando con el ejercicio 4 del Nivel 1: al cliente se le olvidó mencionar un detalle y resulta que los valores en el diccionario original pueden duplicarse, por lo que las claves intercambiadas pueden tener duplicados. En este caso, en el ejercicio anterior imprimías un mensaje de advertencia. Ahora, los valores del diccionario resultante deberán almacenarse como una lista. Ten en cuenta que si es un valor único, no debe ser una lista.  
reverse_dictionary_plus({'x': 'apple', 'y': 'banana', 'z': 'banana'})
{'apple': 'x', 'banana': ['y', 'z']}

```python
def invert_dictionary_with_duplicates(original_dict):
"""This function takes a distionary as input and inverts values with keys (permits duplicate values)"""
  inverted_dict = {}

  for key, value in original_dict.items():
    # Check if the value already exists in the inverted dictionary
    if value in inverted_dict:
      # If it exists and is not a list, convert it to a list
      if not isinstance(inverted_dict[value], list):
        inverted_dict[value] = [inverted_dict[value]]
      # Append the new key to the list
      inverted_dict[value].append(key)
    else:
      # If it doesn't exist, add it as a single key
      inverted_dict[value] = key

  return inverted_dict


# Examples 
original_dict = {"a": 1, "b": 2, "c": 1, "d": 3, "e": 2}
inverted_dict = invert_dictionary_with_duplicates(original_dict)
print("Original dictionary:", original_dict)
print("Inverted dictionary with duplicates:", inverted_dict)

original_dict=({'x': 'apple', 'y': 'banana', 'z': 'banana'})
inverted_dict = invert_dictionary_with_duplicates(original_dict)
print("Original dictionary:", original_dict)
print("Inverted dictionary with duplicates:", inverted_dict)
```

**Ejercicio 2**  
Conversión de tipos de datos  
El cliente recibe una lista de datos y necesita generar dos listas: una que contenga todos los elementos que pudieron convertirse en flotantes y otra con los elementos que no se pudieron convertir.
Ejemplo de la lista que recibe el cliente:
['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']]



```python
def separar_flotantes(datos):
    """This function takes a list as input and returns 2 dictionaries, one for floats and another for NOT floats  """

    lista_convertibles = []
    lista_no_convertibles = []

    def es_convertible_a_float(elemento):
        """This function takes a value and checks if it convertible to float """

        try:
            float(elemento)
            return True
        except (ValueError, TypeError):
            return False

    # Goes through the list and all its elements
    for elemento in datos:
        if isinstance(elemento, (list, tuple)): #checks if the element has type of list or tuple
        #For lists o tuples
            for subelemento in elemento:
                if es_convertible_a_float(subelemento):
                    lista_convertibles.append(subelemento)
                else:
                    lista_no_convertibles.append(subelemento)
        #For NOT lists o tuples
        else:
            if es_convertible_a_float(elemento):
                lista_convertibles.append(elemento)
            else:
                lista_no_convertibles.append(elemento)

    return lista_convertibles, lista_no_convertibles

# Examples 
datos = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']]
convertibles, no_convertibles = separar_flotantes(datos)

print("Lista de elementos convertibles a float:", convertibles)
print("Lista de elementos no convertibles a float:", no_convertibles)

```

Contador y ordenador de palabras en un texto  
Al cliente le gustó el contador de palabras, pero ahora quiere leer archivos TXT y calcular la frecuencia de cada palabra, ordenadas dentro de las entradas habituales del diccionario según la letra con la que comienzan. Es decir, las claves deben ir de la A a la Z, y dentro de la A deben ordenarse de la A a la Z.
Por ejemplo, para el archivo "tu_me_quieres_blanca.txt", la salida esperada sería:
[Proporcionar salida esperada aquí, quizás una lista o un diccionario con las frecuencias de las palabras ordenadas por letra.]

```python
import string
from collections import defaultdict

def leer():
    """ Function to read the file and return its content"""

    archivo = open(r"C:\Users\lasso\Downloads\tu_me_quieres_blanca (1).txt", 'r', encoding='utf-8')
    texto = archivo.read()
    archivo.close()
    return texto

def count_words_by_letter():
    """Function to clean text, count word frequencies, and sort them by starting letter"""
    
    # Read the song text from the file
    song_text = leer()

    # Remove punctuation and convert to lowercase
    custom_punctuation = string.punctuation + "¡" + "!"  # Adding inverted exclamation marks
    table = str.maketrans("", "", custom_punctuation)
    cleaned_text = song_text.translate(table).lower()

    # Split the cleaned text into words
    words = cleaned_text.split()

    # Dictionary to store words grouped by their starting letter
    letter_dict = defaultdict(lambda: defaultdict(int))

    # Count word frequencies and group by the first letter
    for word in words:
        if word:  # Only process non-empty words
            first_letter = word[0]
            letter_dict[first_letter][word] += 1

    # Sort the dictionary by letter (A-Z) and then by word (A-Z) within each letter
    sorted_letter_dict = {}
    for letter in sorted(letter_dict.keys()):
        sorted_letter_dict[letter] = dict(sorted(letter_dict[letter].items()))

    # Return the sorted dictionary
    return sorted_letter_dict

# Get the word frequency count, grouped by letter
sorted_word_counts = count_words_by_letter()

# Print the result
for letter, words in sorted_word_counts.items():
    print(f"'{letter}': {words}")
```
